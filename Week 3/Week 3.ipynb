```python
import pandas as pd, os
from cwd import getPath
from IPython.display import display

os.chdir(getPath())

earthquake_data = pd.read_csv("parsed.csv")
ring_of_fire_loc = ["Alaska", "Antartica", "Bolivia", "California", "Canada", "Chile", "Costa Rica", "Ecuador", "Figi", "Guatemala", 
"Indonesia", "Japan", "Kermadec Islands", "Mexico", "New Zealand", "Peru", "Philippines", "Russia", "Taiwan", "Tonga", "Washington"]

#1) If parsed place is Japan and magType is mb get the 95th percentile of the magnitudes
jp_percentile = earthquake_data[(earthquake_data["parsed_place"]=="Japan") & (earthquake_data["magType"]=="mb")]["mag"].quantile(q=0.95)

#2) If data is earthquake and place is Indonesia, normalize the counts of values in the Tsunami column. (aka the percentage of data they represent.) Then get the data that represents 1
#  (which happens to have an index of 1)
eq_tsu_indonesia = earthquake_data[(earthquake_data["type"]=="earthquake") & (earthquake_data["parsed_place"]=="Indonesia")]["tsunami"].value_counts(normalize=True)[1]

#3) If the place is Nevada and the data is of earthquakes get the list and provide some descriptive statstics, which contains summary statistics. 
nevada_eq = earthquake_data[(earthquake_data["parsed_place"]=="Nevada") & (earthquake_data["type"]=="earthquake")].describe()

#4) Goes through each row in Parsed Place and checks if it is in the ring of fire. It'll return True if it is or False if it isn't. These series of values are then added as a column in 
#the data frame
earthquake_data["Ring of Fire"] = earthquake_data["parsed_place"].apply(lambda x: x in ring_of_fire_loc)

#5) For all earthquakes, count the values in the Ring of Fire column. This returns the total count for True (meaning it is) and False (meaning it isn't)
eq_ring_of_fire_count = earthquake_data[(earthquake_data["type"]=="earthquake")]["Ring of Fire"].value_counts()

#6) For all tsunamis, count of the values in the Ring of Fire column, then return the first row (which are the True values). 
tsu_ring_of_fire_count = earthquake_data[earthquake_data["tsunami"]==1]["Ring of Fire"].value_counts()[0]

print(f'''95th Percentile of Magnitudes in Japan: {jp_percentile} \n 
Percentage of Earthquake and Tsunami combos in Indonesia: {eq_tsu_indonesia} \n
Number of Earthquakes in the Ring of Fire: {eq_ring_of_fire_count[0]} | Number of outside: {eq_ring_of_fire_count[1]} \n
Number of Tsunamis in the Ring of Fire: {tsu_ring_of_fire_count}''')
```
