```python
sunspots = quandl.get("SIDC/SUNSPOTS_A")

print("Head:")
display(sunspots.head())

print("Tail: ")
display(sunspots.tail())

last_date = sunspots.index[-1]
print("Last Date:")
display(pd.DataFrame(sunspots.loc[last_date]))

print("Date String Query: ")
display(sunspots[:"20230107"])

#Show any values greater than the average for the Number of Observations
avg_observation = sunspots["Number of Observations"].mean()
print(f"More than average observations | Mean: {avg_observation}")
display(sunspots[sunspots["Number of Observations"] > avg_observation])

#Show any values greater than the average for the Yearly Mean Total Sunspot Number
yearly_mean = sunspots["Yearly Mean Total Sunspot Number"].mean()
print(f"More than average sunspots | Mean: {yearly_mean}")
display(sunspots[sunspots["Yearly Mean Total Sunspot Number"] > yearly_mean])
```

```python
print("Describe Function")
display(sunspots.describe())

print(f'Observation Count: {sunspots["Number of Observations"].count()}')

#Get the stats and put them in a DF with the stat as a label (since they return a series). Transpose the resulting DF so I can concat them. 
#Note: For some reason, despite DropNA being TRUE by default, I keep getting NaN as my mode. 
agg_stats = pd.concat([pd.DataFrame(sunspots.mad(), columns=["MAD"]).transpose(), 
        pd.DataFrame(sunspots.mean(), columns=["Mean"]).transpose(), 
        pd.DataFrame(sunspots.median(), columns=["Median"]).transpose(), 
        pd.DataFrame(sunspots.max(), columns=["Max"]).transpose(), 
        pd.DataFrame(sunspots.min(), columns=["Min"]).transpose(), 
        pd.DataFrame(sunspots.mode(), columns=["Mode"]).transpose(), 
        pd.DataFrame(sunspots.std(), columns=["Standard Deviation"]).transpose(), 
        pd.DataFrame(sunspots.var(), columns=["Variance"]).transpose(), 
        pd.DataFrame(sunspots.skew(), columns=["Skew"]).transpose()])
agg_stats.drop(axis=1, labels=0, inplace=True)
display(agg_stats)
```

```python
#I'm using the random module here to select random choices from the list, and the NumPY random module for random numbers.
random.seed(1)
np.random.seed(1) 
weather = ["cold", "hot"]
food = ["icecream", "chocolate", "soup"]
#For the weather values select 7 unique values from the weather list (each item repeated 4 times)
#For the food values select 7 unique values from the food list (each item repeated 3 times)
#Generate random prices and numbers
rng_df = pd.DataFrame({'Weather' : random.sample(weather, counts=[4,4], k=7),
   'Food' : random.sample(food, counts=[3,3,3], k=7),
   'Price' : 10 * rand(7), 'Number' : randint(1, 9, size=7)})

rng_df_weather_group = rng_df.groupby("Weather")
rng_df_group_hot = rng_df_weather_group.get_group('hot')
rng_df_group_cold = rng_df_weather_group.get_group('cold')

display(rng_df_group_hot.head(1),rng_df_group_hot.tail(1))
print(rng_df_group_hot.mean(numeric_only=True))

display(rng_df_group_cold.head(1), rng_df_group_cold.tail(1))
print(rng_df_group_cold.mean(numeric_only=True))

rng_df_food_group = rng_df.groupby("Food")
rng_df_group_icecream = rng_df_food_group.get_group("icecream")
rng_df_group_chocolate = rng_df_food_group.get_group("chocolate")
rng_df_group_soup = rng_df_food_group.get_group("soup")


agg_func = {'Price': ['mean', 'median'], 'Number': ['mean', 'median']}

print("Icecream: ")
display(rng_df_group_icecream.agg(agg_func))

print("Chocolate")
display(rng_df_group_chocolate.agg(agg_func))

print("Soup")
display(rng_df_group_soup.agg(agg_func))
```
